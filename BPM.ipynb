```python
import numpy as np
from matplotlib import pyplot as plt
from numpy import loadtxt

%matplotlib inline
```

```python
( _, temp, 
 magX, magY, magZ, magMean, magMeanLength, 
 mat1MagX, mat1MagY, mat1MagZ, mat1MagMean, 
 mat2MagX, mat2MagY, mat2MagZ, mat2MagMean, 
 mat1MagMeanLength, mat2MagMeanLength) = loadtxt("../output", unpack=True)
```

```python
temp.shape
```

```python
plt.figure(figsize=(16,9))

plot1 = plt.plot(temp, magX, 'r--', linewidth = 2, label = '$\hat{m}_{x}$')
plot2 = plt.plot(temp, magY, 'b--', linewidth = 2, label = '$\hat{m}_{y}$')
plot3 = plt.plot(temp, magZ, 'g--', linewidth = 2, label = '$\hat{m}_{z}$')

plt.legend(prop = {'size' : 30})
plt.ylim(-0.2,1.2)
plt.xlim(0, 2000)


plt.xlabel('Temperature', fontsize = 20)
plt.ylabel('Magnetization', fontsize = 20)
plt.title('Unit vector direction - Magnetisation', fontsize = 20)
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(16,9))

plot1 = plt.plot(temp, magMean, 'g--', linewidth = 2, label = '$|m|$')

plt.legend(prop = {'size' : 30})
plt.ylim(0, 1)
plt.xlim(0, 2000)


plt.xlabel('Temperature', fontsize = 20)
plt.ylabel('Magnetization', fontsize = 20)
plt.title('Normalized length of the magnetization', fontsize = 20)
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(16,9))

plot1 = plt.plot(temp, magMeanLength, 'g--', linewidth = 2, label = '$<|m|> $')

plt.legend(prop = {'size' : 30})
plt.ylim(0, 1)
plt.xlim(0, 2000)


plt.xlabel('Temperature', fontsize = 20)
plt.ylabel('Magnetization', fontsize = 20)
plt.title('Normalized Magnetization Length', fontsize = 20)
plt.grid(True)
plt.tight_layout()

plt.show()
```
